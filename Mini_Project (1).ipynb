```python
avg_firstdose = data["First Dose Administered"].astype("float").mean(axis = 0)
print("Average of First Dose:", avg_firstdose)
```

```python
data["First Dose Administered"].fillna(value = avg_firstdose, inplace=True)
data
```

```python
avg_seconddose = data["Second Dose Administered"].astype("float").mean(axis = 0)
print("Average of Second Dose:", avg_seconddose)
```

```python
data["Second Dose Administered"].fillna(value = avg_seconddose, inplace = True)
data
```

```python
# 2. Number of persons state wise vaccinated for first dose in India
first_dose = data.groupby('State')[['First Dose Administered']].sum()
first_dose
```

```python
# 3. Number of persons state wise vaccinated for Second dose in India
Second_dose = data.groupby('State')[['Second Dose Administered']].sum()
Second_dose
```

```python
# 4. Number of Males vaccinated
male = data["Male(Individuals Vaccinated)"].sum()
print("The total number of male individuals vaccinated are", int(male))
```

```python
# 5. Number of Females vaccinated
female = data["Female(Individuals Vaccinated)"].sum()
print("The total number of female individuals vaccinated are", int(female))
```

```python
transgender = data["Transgender(Individuals Vaccinated)"].sum()
print("The total number of Transgender individuals vaccinated are", int(transgender))
```

```python
statewise_data = data.groupby('State')[['Total Doses Administered', 'Total Individuals Vaccinated']].sum()
```

```python
statewise_data.plot(kind='bar', figsize=(12, 6))
plt.title('COVID-19 Vaccine Data - Statewise')
plt.xlabel('State')
plt.ylabel('Number of Doses')
plt.xticks(rotation=45)
plt.legend(['Doses Administered', 'Doses Individuals Vaccinated'])
plt.show()
```
